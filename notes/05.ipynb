```python
def apply_twice(f, x):
    return f(f(x))

def square(x):
    return x * x
```

```python
apply_twice(square, 3)
```

```python
# Example: Repeat

def repeat(f, x):
    while f(x) != x:
        x = f(x)
    return x

def g(y):
    return (y + 5) // 3

result = repeat(g, 5)
```

```python
result
```

```python
def make_adder(n):
    """Return a function that takes one argument k and returns k + n.

    >>> add_three = make_adder(3)
    >>> add_three(4)
    7
    """
    def adder(k):
        return k + n
    return adder

add_three = make_adder(3)
add_three(4)
```
